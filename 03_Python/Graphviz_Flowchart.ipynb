```python
from graphviz import Digraph
```

## Criando e formatando o gráfico

```python
StdMsc = Digraph('Study Music', filename='StudyMusicFlowchart', format='pdf',
               graph_attr={
                'rankdir':'LR',
                'penwidth':'1.0'
               },
               node_attr={
                   'shape':'box',
                   'fontname':'Helvetica',
                   'fontsize':'12'
               },
               edge_attr={
                   'arrowsize':'1.0'
               })
```

## Listagem de pessoas

```python
StdMsc.node('rela',
            'Relaxamento \n do Corpo')

StdMsc.node('aque1',
            'Aquecimento \n com bocal \n acoplado')

StdMsc.node('aque2',
            'Aquecimento \n com bocal não \n acoplado')

StdMsc.node('tecn',
            'Estudos \n técnicos')

StdMsc.node('esca',
            'Escalas')

StdMsc.node('repe',
            'Estudo de \n Repertório')

StdMsc.node('impr',
            'Estudos de \n improvisação')
```

## Definição de Vínculos

```python
StdMsc.edge('rela','aque1')

StdMsc.edge('rela','aque2')

StdMsc.edge('aque1','tecn')

StdMsc.edge('aque2','tecn')

StdMsc.edge('tecn','esca')

StdMsc.edge('repe','esca',dir='back')

StdMsc.edge('impr','repe',dir='back')
```

## Restrições

```python
with StdMsc.subgraph() as s:
    s.attr(rank='same')
    s.node('tecn')
    s.node('esca')
```
