```python
from matplotlib.pyplot import cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
nx, ny = 1.0, 2.5

x = np.arange(0,5, nx)
y = np.arange(40,60, ny)
X, Y = np.meshgrid(x, y)

dy = 10 - Y/5
dx = np.ones(dy.shape)

# Normalize the arrows:
dy = dy / np.sqrt(dy**2 + dx**2);
dx = dx / np.sqrt(dy**2 + dx**2);


plot = plt.figure()
plt.title('Slope field with normalized arrows')
plt.quiver(X, Y, dx, dy,color="lightblue", scale=10, headwidth=0)
```
